```python
import webbrowser

lastIndex = "1426" # เริ่ม -1 # Update 'lastIndex' ให้เริ่มค้นหา

for index, row in df.iterrows():
    value1 = row['เลขทะเบียน']
    value2 = row['ชื่อบริษัทภาษาไทย']
    value3 = row['ข้อมูลสำหรับการติดต่อ']
    value4 = row['เบอร์โทร']
    value5 = row['Website']
    value6 = row['ที่ตั้ง']
    value7 = row['ที่มา']
    
    if (index <= int(lastIndex)): continue
#     if (index <= int(lastIndex)): print("You already did:",index, "-", value1, "-", value2, "-", value3, "-", value4); continue
    
    print(f"\n")
    print(f"{index}: {value1} ---{value3} ---{value4} ---{value5} ---{value2}")
    print(f"         {value6}")
    print(f"         {value7}")
    company_search = value2.replace(' ', '+')
    url = f'https://www.google.com/search?q={company_search}+ติดต่อ'
    print(f"         {url}")
#     webbrowser.open(url)
    
    x = input('เบอร์โทร:')
    if x == '-' : break
    df.at[index, 'เบอร์โทร'] = f"'" + x.replace(' ', '').replace('-', '').replace('(', '').replace(')', '').replace('+66', '0')
#     value3 = x.replace(' ', '').str.replace('+66', '0')
#     str = str + f"\n# Index: {index}\nresult.loc[result['ชื่อบริษัทภาษาไทย'] == '{value1}', 'เบอร์โทร'] = '{value3}'\n"
    
    y = input('Website:')
#     if y == '-' : break
    df.at[index, 'Website'] = y
#     value4 = y
#     str = str + f"result.loc[result['ชื่อบริษัทภาษาไทย'] == '{value1}', 'Website'] = '{value4}'\n"
    
    lastIndex = index

df.loc[df['เบอร์โทร'] == "'", 'เบอร์โทร'] = np.nan

print(f"\nBreak at Index: {lastIndex}\n")

file_name = 'FilterAndSearchData.csv'
file_path = '/Users/bew/Downloads/DataForThai_jupyter/'

# Combine the file path and file namehttps://www.dataforthai.com/company/0105531008413/
full_path = os.path.join(file_path, file_name)

df.to_csv(full_path, sep=',', index=False, header=True, encoding='utf-8')

print('fill done')
```
