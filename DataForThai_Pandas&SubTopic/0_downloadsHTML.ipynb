```python
import pandas as pd
import requests

# URL of the webpage containing the table
url = 'https://www.dataforthai.com/business/objective'

# Use headers to mimic a request from a regular web browser
headers = {
    'User-Agent': 'PostmanRuntime/7.37.3',
    'Accept': '*/*',
    'Cookie': 'PHPSESSID=j0ufdk464g41pid5am3932l3hu'
}

# Send a GET request to the webpage
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Read the HTML content using pandas
    tables = pd.read_html(response.text)
    
    # Assuming you want the first table (index 0)
    df = tables[1]
    
    # Display the DataFrame
    print(df)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
df = df.sort_values(by='รหัส', ascending=True)
df['รหัส'] = df['รหัส'].astype(str)
# Function to add '0' to strings of length 4
def add_zero(s):
    if len(s) == 4:
        return '0' + s
    else:
        return s

# Applying the function to the DataFrame
df['รหัส'] = df['รหัส'].apply(lambda x: add_zero(x))
df
```
